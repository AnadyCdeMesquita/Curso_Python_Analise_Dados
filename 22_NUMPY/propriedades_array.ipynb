```python
import numpy as np

array = np.array([50, 80, 100, 120, 150])
print(array)
print('Caracteristica do ARRAY')
print('TAMANHO:', array.size)
print('FORMATO:', array.shape)
print('DIMENSÃO:', array.ndim)
print('TIPO:',array.dtype)

matriz = np.array([[10, 20, 30], [40, 50, 60]])
print('*' * 50)
print(matriz)
print('Caracteristica do MATRIZ')
print('TAMANHO:', matriz.size)
print('FORMATO:', matriz.shape)
print('DIMENSÃO:', matriz.ndim)
print('TIPO:',matriz.dtype)

```
