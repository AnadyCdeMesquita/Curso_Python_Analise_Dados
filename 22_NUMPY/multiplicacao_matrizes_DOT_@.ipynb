```python
import numpy as np

# A multiplicação de duas matrizes \(2\times2\) (\(A \times B\))
# é realizada multiplicando as linhas da primeira matriz pelas colunas da segunda,
# somando os produtos dos elementos correspondentes.

a = np.array([[1, 2],
              [4, 5]])
b = np.array([[7, 8],
             [10, 11]])

resultado = a @ b
print(resultado)

resultado2 = np.dot(a, b)

```
