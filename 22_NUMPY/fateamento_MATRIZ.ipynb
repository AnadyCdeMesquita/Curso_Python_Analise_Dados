```python
import numpy as np

matriz = np.array([[10, 20, 30],
                   [40,50,60], 
                   [70,80,90]])

print(matriz[0, 0:2]) #pega o numero do elemento(linha) no array e depois pega os indices
print(matriz[0:3, 2])#pega os índices, depois pega só a coluna. 
#Seleciona só a quantidade de índices desejados.
print(matriz[1:3, :2])
```
