```python
import numpy as np

notas = np.array([10.0, 5.0, 8.5, 9.0, 7.5, 6.6, 2.5])
notas1 = np.sort(notas)
print(notas1)

soma = np.sum(notas1)
print(soma)
maxima = np.max(notas1)
print(maxima)
minima = np.min(notas1)
print(minima)
media = np.mean(notas1)
print(media)
Desvio_Padrao = np.std(notas1)
print(Desvio_Padrao)
mediana = np.median(notas1)
print(mediana)
variancia = np.var(notas1)
print(variancia)
```
