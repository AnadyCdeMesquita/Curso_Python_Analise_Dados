```python
import numpy as np

dados = np.array([10, 20, 30, 40, 50, 60])
print(dados)

dados[5] = 90
print('novosdados', dados)

#alterando varios elementos

dados[0:3] = [15, 25, 35]
print('ARRAYS', dados)

novo_array = np.array([10, 5.5, 11, 16]) # transformou todos em float
print('Array com tipos diferentes', novo_array)
print('Array com tipos diferentes', novo_array.dtype)

novo_array2 = np.array([10, 'Anady', 11, 16]) # transformou todos em strings
print('Array com tipos diferentes', novo_array2)
print('Array com tipos diferentes', novo_array2.dtype)
```
