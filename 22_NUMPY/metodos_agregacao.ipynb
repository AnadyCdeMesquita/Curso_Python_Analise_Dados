```python
import numpy as np
# Para somar todos os números do array, utilize np.sum(array) ou o método array.sum()
dados = np.array([10, 20, 30, 40])
soma = np.sum(dados) # ou dados.sum()
print(soma)

media= dados.mean()
print(media)

maximo= dados.max()
print(maximo)

minimo= dados.min()
print(minimo)

desvio_padrao = dados.std()
print(desvio_padrao)
```
