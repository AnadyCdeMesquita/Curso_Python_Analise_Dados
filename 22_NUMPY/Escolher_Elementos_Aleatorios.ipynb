```python
import numpy as np

lista_objetos = ['anady', 'fernando', 'soph', 'lara', 'juju']

selecionar = np.random.choice(lista_objetos) # seleciona apenas 1
print(selecionar)

selecionar2 = np.random.choice(lista_objetos, 4) # seleciona a quantidade de objetos que podem se repetir
print(selecionar2)

selecionar3 = np.random.choice(lista_objetos, 4, replace=False) # seleciona a quantidade de objetos e para não se repetir, coloca-se replace= False
print(selecionar3)
```
