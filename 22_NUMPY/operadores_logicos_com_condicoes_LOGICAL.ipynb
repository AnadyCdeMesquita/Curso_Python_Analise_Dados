```python
import numpy as np

dados = np.array([10, 20, 30, 40, 50, 60])

cond = np.logical_and(dados > 20,  dados >=50)
print(cond)
print(dados[cond])

cond2 = np.logical_or(dados < 20, dados <=40)
print(cond2)
print(dados[cond2])

cond3 = np.logical_not(dados < 30)
print(cond3)
print(dados[cond3])
```
